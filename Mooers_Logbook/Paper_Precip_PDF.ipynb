```python
path_to_file = '/fast/gmooers/Real_Geography_Manuscript/Preprocessed_Data/PRECIP_ANNUAL/full_physics_essentials_valid_month02_targets.nc'
SPCAM5_ds = xr.open_dataset(path_to_file)

lons = SPCAM5_ds.lon.values
lats = SPCAM5_ds.lat.values

lats = lats[::144][:96]
lons = lons[:144]

path_to_file = '/fast/gmooers/Real_Geography_Manuscript/Models/Paper_Full_2018.nc'
SHERPA_ds = xr.open_dataset(path_to_file)

path_to_file = '/fast/gmooers/Real_Geography_Manuscript/Models/Annual_Precip_Constrained.nc'
CONSTRAINED_ds = xr.open_dataset(path_to_file)

path_to_file = '/fast/gmooers/Real_Geography_Manuscript/Models/Paper_Full_2018_Primative.nc'
UNCONSTRAINED_ds = xr.open_dataset(path_to_file)

path_to_file = '/fast/gmooers/Real_Geography_Manuscript/Models/CAM_12_Months_Correct.nc'
CAM_ds = xr.open_dataset(path_to_file)
```

```python
precip_SPCAM5_ds = np.squeeze(SPCAM5_ds.targets.values)*86400000.

precip_SHERPA_ds = np.squeeze(SHERPA_ds.Prediction.values)*86400000.

precip_CONSTRAINED_ds = np.squeeze(CONSTRAINED_ds.Prediction.values)*86400000.

precip_UNCONSTRAINED_ds = np.squeeze(UNCONSTRAINED_ds.Prediction.values)*86400000.

precip_cam_ds = (CAM_ds.PRECT.values)*86400000.

times = SPCAM5_ds.time.values

Xvar, Yvar = np.meshgrid(lons, lats)

x = 144
y = 96
z = 30
t = int(len(precip_SPCAM5_ds)/(x*y))
print(t)
```

```python
reconstructed_SPCAM5 = np.zeros(shape=(t, y, x))
reconstructed_SHERPA = np.zeros(shape=(t, y, x))
reconstructed_CONSTRAINED = np.zeros(shape=(t, y, x))
reconstructed_UNCONSTRAINED = np.zeros(shape=(t, y, x))
```

```python
count = 0
for i in range(t):
    for j in range(y):
        for k in range(x):
            reconstructed_SPCAM5[i,j,k] = precip_SPCAM5_ds[count]
            reconstructed_SHERPA[i,j,k] = precip_SHERPA_ds[count]
            reconstructed_CONSTRAINED[i,j,k] = precip_CONSTRAINED_ds[count]
            reconstructed_UNCONSTRAINED[i,j,k] = precip_UNCONSTRAINED_ds[count]
            count = count + 1
                
```

```python
def array_generator(min_val, max_val, percentage):
    listing = []
    for i in range(300):
        if i == 0:
            temp_val = min_val
            listing.append(temp_val)
        else:
            if temp_val < max_val:
                temp_val = temp_val*percentage
                listing.append(temp_val)
        
    print(len(listing))
    return np.array(listing)
#array_generator(0.03, maxi, 1.07)
```
